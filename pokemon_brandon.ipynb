```python
#Cargamos el drataFrame de los alumnos
url = '/content/Pokemon.csv'
df = pd.read_csv(url, sep=",")

#Imprimiendo los datos
pd.read_csv(url)
```

```python
#Identificando la relaciónes entra las variables
#Con el propostido de encontra algun patros y averigual si se puede identificar el tipo de pokemon

y = df["Sp. Atk"]
x = df["Sp. Def"]
#Labels
plt.title('Identificando el tipo de pokemon la relación entre HP y Attack')
plt.xlabel('Generation')
plt.ylabel('Legendary')
plt.scatter(x, y, color = '#88c999')

plt.show()

# Simpre tener la lista de las columnas para optimizar el tiempo
# Name	Type 1	Type 2	Total	HP	Attack	Defense	Sp. Atk	Sp. Def	Speed	Generation	Legendary
```

```python
df[['Sp. Atk','Sp. Def']].corr()
```

**¿Podemos sacar una conclusión sobre la relación entre la variable ataque y defensa?**

R: Segun el grafico de dispersion, podemos notar a simple vista que hay un aumento en el Sp. Atk a medida que tiene una mejor Sp. Def.

Calculado la Correlacion entre estas dos variables obtenemos un 0.5.. con esto, podemos decir que hay una correlacion moderada.

Tambien, concluir en que no se puede afirmar que estas variables tengan una relacion directa pero si moderada.

```python
#Histograma de la variable total

plt.figure(figsize=(5,5))
plt.hist(x = df['Total'], bins = 20)
plt.xlabel("Total")
plt.ylabel("Frecuencia")
plt.title("Variable: Total (suma de sus poderes)")
plt.show()
```

```python
df['Total'].mode()
```

```python
df['Total'].median()
```

```python
df['Total'].skew()
```

```python
df['Total'].mean() # media
```

**¿Que podemos intepretar de este histograma? ¿Cual es su asimetria y que tipo de distribucion tiene el histograma?**

**Interpretación, Analisis de la media y moda**

R: 1. Primero podemos intepretar que hay una gran cantidad de frecuencia de datos entre los 400 y 500 de la variable total.
   
   2. Sacando la mediana la cual apunta al 50% de la gran cantidad de datos podemos obtener el valor de 450 la cual podemos decir que ese es su punto medio.
   
   3. En la media o promedio obtenemos 435.1025 de los datos.
   
   4. Por ultimo sacando la moda de la variable Total, podemos decir que el valor que mas se repite es el 600.

      
**Tipo de distribución**

R: Podemos decir que segun la media y mediana el histograma no sigue una distribucion normal ya que no tiene la forma inusual de campana, sino una distribucion sesgada.

```python
# Name	Type 1	Type 2	Total	HP	Attack	Defense	Sp. Atk	Sp. Def	Speed	Generation	Legendary

#Grafico de caja
sns.catplot(x="Type 1", y="Total", data=df, kind="box", height=8, aspect=2)
plt.show() # todo lo que esta afuera son datos atipicos (estudiar este grafico)
```
